## Part 1: Enhanced Expense Entry (15 minutes)

### 🔄 From One to Many

Transform your Week 2 single-expense tracker to handle multiple expenses with proper validation:

```python
print("=== Enhanced Finance Tracker v0.3 ===")
print("Week 3: Loop-Powered Edition")
print()

# User setup (enhanced from Week 2)
name = input("Enter your name: ")

# Add budget validation using while loop
monthly_budget = 0
while monthly_budget < 100 or monthly_budget > 10000:
    try:
        budget_input = input("Enter monthly budget ($100-$10,000): $")
        monthly_budget = float(budget_input)

        if monthly_budget < 100:
            print("Budget too low! Minimum $100")
        elif monthly_budget > 10000:
            print("Budget seems high! Maximum $10,000")
    except ValueError:
        print("Please enter a valid number!")

print(f"\nWelcome {name}! Monthly budget: ${monthly_budget:.2f}")

# ========== MULTI-EXPENSE ENTRY ==========
expenses = []  # Store all expense amounts
descriptions = []  # Store what each expense was for
total_spent = 0

print("\n--- Enter Your Expenses ---")
print("Type 'done' when finished entering expenses")

# Your task: Complete the multi-expense entry system
# Use sentinel value pattern to collect multiple expenses
# Validate each expense amount (positive numbers only)
# Store both amount and description
# Show running total and budget percentage after each entry

while True:
    expense_input = input("\nExpense amount (or 'done'): $")

    if expense_input.lower() == 'done':
        break
    try:
        expense = float(expense_input)
        if expense <= 0:
            print("Expense must be greater than $0")
            continue

        description = input("What was this expense for: ")
        expenses.append(expense)
        descriptions.append(description)
        total_spent += expense

        print(f"Added: ${expense:.2f} for {description}")

        percent_used = (total_spent / monthly_budget) * 100
        print(f"Running total: ${total_spent:.2f}")
        print(f"Budget used: {percent_used:.1f}%")

    except ValueError:
        print("Please enter a valid number!")



print(f"\n📊 Expense entry complete: {len(expenses)} expenses, ${total_spent:.2f} total")
```

## Part 2: Basic Expense Analysis (15 minutes)

### 📈 Processing Your Transaction Data

Use loops to analyze the expenses you collected:

```python
# Expense Analysis Section
if len(expenses) > 0:
    print("\n📊 Expense Analysis:")

    total_spent = sum(expenses)

    print(f"Total expenses: {len(expenses)}")
    print(f"Total spent: ${total_spent:.2f}")
    print(f"Average expense: ${total_spent/len(expenses):.2f}")

    budget_used = (total_spent / monthly_budget) * 100 if monthly_budget > 0 else 0
    print(f"Budget used: {budget_used:.1f}%")
    print(f"Remaining budget: ${monthly_budget - total_spent:.2f}")

    # Highest & Lowest
    highest_expense = expenses[0]
    lowest_expense = expenses[0]

    for expense in expenses:
        if expense > highest_expense:
            highest_expense = expense
        if expense < lowest_expense:
            lowest_expense = expense

    print(f"\n💰 Expense Range:")
    print(f"Highest expense: ${highest_expense:.2f}")
    print(f"Lowest expense: ${lowest_expense:.2f}")

    # Categorization
    large_count = 0
    medium_count = 0
    small_count = 0

    for expense in expenses:
        percent = (expense / monthly_budget) * 100 if monthly_budget > 0 else 0

        if percent > 20:
            large_count += 1
        elif percent >= 5:
            medium_count += 1
        else:
            small_count += 1

    print(f"\n📊 Expense Categories:")
    print(f"Large expenses (>20% budget): {large_count}")
    print(f"Medium expenses (5-20% budget): {medium_count}")
    print(f"Small expenses (<5% budget): {small_count}")

    print("\n💡 Smart Feedback:")

    if budget_used > 100:
        print("⚠️ You have exceeded your monthly budget!")
    elif budget_used > 80:
        print("⚠️ You are close to your budget limit.")
    else:
        print("✅ Your spending is within the budget.")

    if large_count > 2:
        print("💡 You have many large expenses. Consider reducing big purchases.")
    elif small_count > medium_count:
        print("👍 Most of your expenses are small. Good spending control!")

else:
    print("\n📝 No expenses entered - try adding some transactions!")
```

## Part 3: Multiple Savings Goals Tracker (15 minutes)

### 🎯 Goal Management with Loops

Extend your tracker to handle multiple savings goals:

```python
print("\n=== Savings Goals Tracker ===")

# Goal collection
goal_names = []
goal_targets = []
goal_current = []

# Ask how many goals they want to track
num_goals = 0
while num_goals < 1 or num_goals > 5:  # Reasonable limits
    try:
        num_goals = int(input("How many savings goals? (1-5): "))
        if num_goals < 1:
            print("Must have at least 1 goal!")
        elif num_goals > 5:
            print("Maximum 5 goals for this tracker!")
    except ValueError:
        print("Please enter a valid number!")

# Your task: Collect goal information using for loop
# For each goal (1 to num_goals):
#   - Get goal name
#   - Get target amount (with validation)
#   - Get current amount (with validation)
#   - Store in the three lists
for i in range(num_goals):
    print(f"\n--- Goal {i+1} ---")

    goal_name = input("Enter goal name: ")
    goal_names.append(goal_name)

    while True:
        try:
            target = float(input("Enter target amount: "))
            if target >= 0:
                goal_targets.append(target)
                break
            else:
                print("Amount must be positive!")
        except ValueError:
            print("Please enter a valid number!")

    while True:
        try:
            current = float(input("Enter current amount: "))
            if current >= 0:
                goal_current.append(current)
                break
            else:
                print("Amount must be positive!")
        except ValueError:
            print("Please enter a valid number!")

print(f"\nEntering {num_goals} savings goals:")

for i in range(num_goals):
    print(f"\n--- Goal {i+1} ---")

    # Your goal collection code here
    print(f"Name: {goal_names[i]}")

print(f"\n✅ Saved {len(goal_names)} goals!")
```

```python
# Goal Analysis and Progress Tracking
if len(goal_names) > 0:
    print("\n📊 Goals Progress Analysis:")

    # Your task: Analyze each goal
    # For each goal, calculate:
    #   - Progress percentage
    #   - Amount remaining
    #   - Months to completion (if saving $100/month)

    total_target = 0
    total_current = 0
    lowest_progress = 100
    lowest_index = -1

    for i in range(len(goal_names)):
       progress = (goal_current[i] / goal_targets[i]) * 100 if goal_targets[i] > 0 else 0
       remaining = goal_targets[i] - goal_current[i]
       months_to_complete = remaining / 100 if remaining > 0 else 0

    print(f"\nGoal: {goal_names[i]}")
    print(f"  Progress: {progress:.1f}%")
    print(f"  Amount remaining: ${remaining:.2f}")
    print(f"  Months to completion (at $100/month): {months_to_complete:.1f}")


    total_target += goal_targets[i]
    total_current += goal_current[i]

    if progress < lowest_progress:
            lowest_progress = progress
            lowest_index = i


    overall_progress = (total_current / total_target) * 100 if total_target > 0 else 0
    total_remaining = total_target - total_current

    print(f"\n🎯 Overall Goals Summary:")
    print(f"Total target: ${total_target:.2f}")
    print(f"Total saved: ${total_current:.2f}")
    print(f"Overall progress: {overall_progress:.1f}%")
    print(f"Total remaining: ${total_remaining:.2f}")

    if lowest_index != -1:
     print(f"\n⚠️ Goal needing most attention: {goal_names[lowest_index]}")
     print(f"  Progress: {lowest_progress:.1f}%")
     print("  Advice: Keep saving consistently! Even $100/month will get you there.")


else:
    print("No goals to analyze!")
```

## Part 4: Simple Menu System (10 minutes)

### 🎮 Interactive Program Flow

Create a basic menu that lets users choose what to do:

```python
print("\n=== Interactive Finance Menu ===")


name = input("Enter your name: ")

expenses = []
descriptions = []

goal_names = ["Emergency Fund", "Vacation"]
goal_targets = [1000, 2000]
goal_current = [200, 500]

monthly_budget = 1500


running = True

while running:
    print("\n" + "-"*30)
    print(f"💰 {name}'s Finance Tracker")

    if len(expenses) > 0:
        print(f"📊 Current: {len(expenses)} expenses, ${sum(expenses):.2f} spent")

    print("-"*30)
    print("1. View expense summary")
    print("2. View goals progress")
    print("3. Add new expense")
    print("4. Budget check")
    print("5. Exit")

    choice = input("\nChoose option (1-5): ")


    if choice == "1":
        print("\n--- Expense Summary ---")

        if len(expenses) == 0:
            print("No expenses recorded yet.")
        else:
            for i in range(len(expenses)):
                print(f"{i+1}. ${expenses[i]:.2f} - {descriptions[i]}")

            print(f"Total spent: ${sum(expenses):.2f}")


    elif choice == "2":
        print("\n--- Goals Progress ---")

        if len(goal_names) == 0:
            print("No goals set yet.")
        else:
            for i in range(len(goal_names)):
                percent = (goal_current[i] / goal_targets[i]) * 100 if goal_targets[i] > 0 else 0
                print(f"{goal_names[i]}: ${goal_current[i]:.2f}/ ${goal_targets[i]:.2f} ({percent:.0f}% complete)")


    elif choice == "3":
        print("\n--- Add New Expense ---")

        exp_name = input("Enter expense name: ")

        try:
            amount = float(input("Enter expense amount: "))
        except ValueError:
            print("Invalid amount. Please enter a number.")
            continue

        expenses.append(amount)
        descriptions.append(exp_name)

        print(f"Added: ${amount:.2f} - {exp_name}")


    elif choice == "4":
        print("\n--- Budget Check ---")

        total_spent = sum(expenses)
        remaining_budget = monthly_budget - total_spent
        percent = (total_spent / monthly_budget) * 100 if monthly_budget > 0 else 0

        print(f"Budget: ${monthly_budget:.2f}")
        print(f"Total spent: ${total_spent:.2f}")
        print(f"Remaining budget: ${remaining_budget:.2f}")
        print(f"Budget used: {percent:.1f}%")

        if percent > 90:
            print("⚠️ You're close to reaching your monthly budget!")
        elif percent > 70:
            print("⚠️ Consider cutting back on expenses.")
        else:
            print("✅ Your spending is within the budget.")



    elif choice == "5":
        running = False
        print(f"\nGoodbye {name}!")
        print(f"Final summary: {len(expenses)} expenses, ${sum(expenses):.2f}")
        print("🚀 Your finance tracker is growing stronger each week!")


    else:
        print("Invalid choice! Please choose 1-5.")

print("\n=== Finance Tracker Session Complete ===")
```

```python
# Finance Tracker

name = input("Enter your name: ")
budget = float(input("Enter your total budget: "))

expenses = []
goals = {}


for i in range(2):
    goal_name = input("Enter goal name: ")
    target = float(input("Enter target amount: "))
    current = float(input("Enter current amount: "))
    goals[goal_name] = {"target": target, "current": current}


while True:
    entry = input("Enter expense amount (or 'done'): ")
    if entry.lower() == "done":
        break
    amount = float(entry)
    desc = input("Enter description: ")
    expenses.append((amount, desc))


while True:
    print("\nMenu:")
    print("1. View expense summary")
    print("2. View goals progress")
    print("3. Add expense")
    print("4. Budget check")
    print("5. Exit")

    choice = input("Choose an option: ")

    if choice == "1":
        total = sum(e[0] for e in expenses)
        print(f"Total expenses: {len(expenses)}")
        print(f"Total spent: ${total:.2f}")

    elif choice == "2":
        for name, g in goals.items():
            percent = (g["current"] / g["target"]) * 100
            print(f"{name}: {percent:.1f}% complete")

    elif choice == "3":
        amount = float(input("Enter amount: "))
        desc = input("Enter description: ")
        expenses.append((amount, desc))

    elif choice == "4":
        total = sum(e[0] for e in expenses)
        percent = (total / budget) * 100
        print(f"Total spent: ${total:.2f}")
        print(f"Budget used: {percent:.1f}%")

    elif choice == "5":
        print("Goodbye!")
        break

    else:
        print("Invalid option.")
```
